```python
trainY = train['Survived']
```

```python
row, col = train.shape
for r in range(row):
  if train['Pclass'][r] == 1:
    train['Age'][r] = 38
  elif train['Pclass'][r] == 2:
    train['Age'][r] = 30
  else:
    train['Age'][r] = 25
train.head(5)
```

```python
row, col = test.shape
for r in range(row):
  if test['Pclass'][r] == 1:
    test['Age'][r] = 38
  elif test['Pclass'][r] == 2:
    test['Age'][r] = 30
  else:
    test['Age'][r] = 25
test.head(5)
```

```python
train.drop(['Cabin', 'Name', 'Ticket', 'PassengerId','Survived'], axis = 1, inplace = True)
train.head()
```

```python
test.drop(['Cabin', 'Name', 'Ticket', 'PassengerId'], axis = 1, inplace = True)
test.head()
```

```python
embarked = pd.get_dummies(train['Embarked'])
sex = pd.get_dummies(train['Sex'])
# train.drop(['Embarked', 'Sex'], axis = 1, inplace = True)
train = train.drop(['Embarked', 'Sex'], axis = 1)
```

```python
train = pd.concat([train, sex, embarked], axis = 1)
train.head()
```

```python
embarked2 = pd.get_dummies(test['Embarked'])
sex2 = pd.get_dummies(test['Sex'])
# train.drop(['Embarked', 'Sex'], axis = 1, inplace = True)
test = test.drop(['Embarked', 'Sex'], axis = 1)
```

```python
test = pd.concat([test, sex2, embarked2], axis = 1)
test.head()
```

```python
train_data = train.to_numpy()
row,column = train_data.shape
def centralize(x):
  mean = np.mean(x)
  stddev = np.std(x)
  x = (x - mean)/stddev
  return x

for col in range(column):
  train_data[:, col] = centralize(train_data[:, col])

train_data[:5]
```

```python
test_data = test.to_numpy()
row2,column2 = test_data.shape

def centralize_test(y):
  mean2 = np.mean(y)
  stddev2 = np.std(y)
  y = (y - mean2)/stddev2
  return y

for col in range(column2):
  test_data[:, col] = centralize_test(test_data[:, col])
for i in range(0,row2) :
    for j in range(0,column2) :
        if (np.isnan(test_data[i][j])) :
            test_data[i][j] = -1
test_data[:5]
```

```python

num = 250
def score(x,xt) :
    summ = np.exp(np.matmul(np.transpose(x),xt))
    return summ

total = 0
for i in range(0,num) :

    total += score(train_data[i],test_data[i])

summ = 0
arr = [None]*num
k = 0
for j in range(0,num) :
    summ = (score(train_data[j],test_data[j])/total)

    if (trainY[j] == 0) :
        trainY[j] = -1

    arr[k] = summ
    k+=1

a=0
for p in range(0,num) :
    if (trainY[p] == 1 ) :
        a += arr[p]*trainY[p]

if (a > 0) :
    print(f"Survived\nAccuracy:{(a)*100}%")
else :
    print(f"Not Survived\nAccuracy:{(a)*100}%")
```
